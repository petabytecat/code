```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===== CONFIGURABLE PARAMETERS =====
# Image cropping parameters - adjust these as needed
CROP_TOP = 95      # Pixels to crop from top
CROP_BOTTOM = 195   # Pixels to crop from bottom
CROP_LEFT = 0       # Pixels to crop from left
CROP_RIGHT = 0      # Pixels to crop from right
RESIZE_WIDTH = 100   # Final width after resizing
RESIZE_HEIGHT = 50  # Final height after resizing

# Training hyperparameters
LEARNING_RATE = 0.00025
N_EPISODES = 10000
START_EPSILON = 1.0
FINAL_EPSILON = 0.1
EPSILON_DECAY = (START_EPSILON - FINAL_EPSILON) / 1000000  # Decay over 1M steps
DISCOUNT_FACTOR = 0.99
MEMORY_CAPACITY = 100000
MEMORY_FILL_SIZE = 50000
MINIBATCH_SIZE = 32
TARGET_UPDATE_FREQ = 10000
REWARD_CLIP = True  # Clip rewards to [-1, 1]
# ===================================
# ===== PREPROCESS FUNCTION =====
def preprocess(obs):
    """Convert RGB to grayscale, crop, and resize."""
    gray = np.dot(obs[..., :3], [0.299, 0.587, 0.114])
    cropped = gray[95:195]  # Crop height
    resized = cv2.resize(cropped, dsize=(100, 50), interpolation=cv2.INTER_NEAREST)
    return resized.astype(np.uint8)

# ===== DQN MODEL DEFINITION =====
import torch.nn as nn
import torch.nn.functional as F

class DQN(nn.Module):
    def __init__(self, num_actions, in_channels=4):
        super(DQN, self).__init__()

        # Convolutional layers (standard DQN architecture)
        self.conv1 = nn.Conv2d(in_channels, 32, kernel_size=8, stride=4)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=4, stride=2)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, stride=1)

        # Dynamically calculate linear layer size
        with torch.no_grad():
            dummy = torch.zeros(1, in_channels, RESIZE_HEIGHT, RESIZE_WIDTH)
            dummy = F.relu(self.conv1(dummy))
            dummy = F.relu(self.conv2(dummy))
            dummy = F.relu(self.conv3(dummy))
            self.linear_size = dummy.view(1, -1).size(1)

        self.fc1 = nn.Linear(self.linear_size, 512)
        self.fc2 = nn.Linear(512, num_actions)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(x.size(0), -1)  # Flatten
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# ===== PLAY FUNCTION =====
def play_episodes(model_path, env_id="ALE/Gopher-v5", num_episodes=3, max_steps=10000, delay=0.02):
    """Play episodes using a trained model and show preprocessed input"""
    try:
        env = gym.make(env_id, render_mode="human")
    except gym.error.Error as e:
        print(f"Failed to create environment: {e}")
        print("Ensure you've installed required packages:")
        print("  pip install gymnasium[atari] gymnasium[accept-rom-license]")
        sys.exit(1)
    
    model = DQN(env.action_space.n).to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()

    # Create OpenCV windows
    cv2.namedWindow("Original Frame", cv2.WINDOW_NORMAL)
    cv2.namedWindow("Preprocessed Input", cv2.WINDOW_NORMAL)
    cv2.namedWindow("Network Input (State)", cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Original Frame", 400, 400)
    cv2.resizeWindow("Preprocessed Input", 200, 100)
    cv2.resizeWindow("Network Input (State)", 400, 200)

    for episode in range(num_episodes):
        obs, info = env.reset()
        frame = preprocess(obs)
        frame_stack = deque(maxlen=4)
        for _ in range(4):
            frame_stack.append(frame)
        state = np.stack(frame_stack, axis=0)

        total_reward = 0

        for step in range(max_steps):
            # Show original frame
            original_display = cv2.cvtColor(obs, cv2.COLOR_RGB2BGR)
            cv2.imshow("Original Frame", original_display)
            
            # Show latest preprocessed frame
            cv2.imshow("Preprocessed Input", frame_stack[-1])
            
            # Show full state input (4 frames)
            state_display = np.zeros((RESIZE_HEIGHT*2, RESIZE_WIDTH*2), dtype=np.uint8)
            state_display[0:RESIZE_HEIGHT, 0:RESIZE_WIDTH] = frame_stack[0]  # Frame t-3
            state_display[0:RESIZE_HEIGHT, RESIZE_WIDTH:] = frame_stack[1]  # Frame t-2
            state_display[RESIZE_HEIGHT:, 0:RESIZE_WIDTH] = frame_stack[2]  # Frame t-1
            state_display[RESIZE_HEIGHT:, RESIZE_WIDTH:] = frame_stack[3]    # Frame t
            state_display = cv2.resize(state_display, (400, 200), interpolation=cv2.INTER_NEAREST)
            cv2.imshow("Network Input (State)", state_display)
            
            # Process keypress
            key = cv2.waitKey(1)
            if key == ord('q'):
                env.close()
                cv2.destroyAllWindows()
                return

            # Get action from model
            state_tensor = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0) / 255.0
            with torch.no_grad():
                q_values = model(state_tensor)
            action = q_values.argmax().item()

            next_obs, reward, done, truncated, _ = env.step(action)
            total_reward += reward

            env.render()
            time.sleep(delay)

            frame = preprocess(next_obs)
            frame_stack.append(frame)
            next_state = np.stack(frame_stack, axis=0)

            if done or truncated:
                break
            state = next_state
            obs = next_obs  # Update for visualization

        print(f"Episode {episode+1}: Total reward: {total_reward}, Steps: {step+1}")

    env.close()
    cv2.destroyAllWindows()

# ===== EXECUTE PLAYBACK =====
if __name__ == "__main__":
    print("Playing with best model...")
    play_episodes("dqn_gopher_best.pth")
```

```python
# defining neural network 
# based on this architecture: https://arxiv.org/pdf/1312.5602
# code written by Claude Sonnet 4

class DQN(nn.Module):
    def __init__(self, actions):
        """
        Deep Q-Network adapted for 100x50x4 input.
        
        Args:
            num_actions (int): Number of valid actions
        """
        super(DQN, self).__init__()
        
        # first convolutional layer: 16 filters of 8x8 with stride 4
        self.conv1 = nn.Conv2d(in_channels=4, out_channels=16, kernel_size=8, stride=4)
        
        # second convolutional layer: 32 filters of 4x4 with stride 2
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=4, stride=2)
        
        # calculate the size of flattened features after convolutions
        # input: 100x50x4 -> Conv1: 24x11x16 -> Conv2: 11x4x32 = 1408 features
        self.fc1 = nn.Linear(32 * 11 * 4, 256)
        
        # output layer: one output for each possible action
        self.fc2 = nn.Linear(256, actions)
        
    def forward(self, x):
        """
        Forward pass through the network.
        
        Args:
            x (torch.Tensor): Input tensor of shape (batch_size, 4, 100, 50)
                             representing 4 stacked 100x50 grayscale frames
        
        Returns:
            torch.Tensor: Q-values for each action, shape (batch_size)
        """
        # first conv layer with ReLU activation
        x = F.relu(self.conv1(x))
        
        # second conv layer with ReLU activation
        x = F.relu(self.conv2(x))
        
        # flatten for fully connected layers
        x = x.view(x.size(0), -1)
        
        # fully connected layer with ReLU activation
        x = F.relu(self.fc1(x))
        
        # output layer (no activation - linear)
        x = self.fc2(x)
        
        return x

# example input: batch of 32 preprocessed frames
# batch_size = 32
# input_tensor = torch.randn(batch_size, 4, 100, 50)

# forward pass
# q_values = model(input_tensor)
# print(f"Input shape: {input_tensor.shape}")
# print(f"Output Q-values shape: {q_values.shape}")
```

```python
# replay memory 
class ReplayMemory:
    def __init__(self, capacity):
        self.memory = deque(maxlen=capacity)

    def push(self, transition):
        """Save a transition."""
        self.memory.append(transition)

    def sample(self, batch_size):
        """Sample a random batch of transitions."""
        return random.sample(self.memory, batch_size)

    def __len__(self):
        return len(self.memory)
```

```python
# initialize replay memory
replay_memory = ReplayMemory(
    capacity=capacity,
)

# reset environment and initialize frame stack
obs, info = env.reset()
frame = preprocess(obs)

frame_stack = deque(maxlen=4)
for _ in range(4):
    frame_stack.append(frame)
state = np.stack(frame_stack, axis=0)

# pre-fill replay memory
for _ in tqdm(range(memory_fill_size), desc="Pre-filling replay memory"):
    action = env.action_space.sample()
    next_obs, reward, done, truncated, info = env.step(action)

    next_frame = preprocess(next_obs)
    frame_stack.append(next_frame)
    next_state = np.stack(frame_stack, axis=0)

    replay_memory.push((state, action, reward, next_state, done))
    state = next_state

    if done or truncated:
        obs, info = env.reset()
        frame = preprocess(obs)
        frame_stack.clear()
        for _ in range(4):
            frame_stack.append(frame)
        state = np.stack(frame_stack, axis=0)
```

```python
# Initialize networks and optimizer
online_net = DQN(env.action_space.n).to(device)
target_net = DQN(env.action_space.n).to(device)
target_net.load_state_dict(online_net.state_dict())
target_net.eval()

# Optimizer with paper hyperparameters
optimizer = optim.RMSprop(
    online_net.parameters(),
    lr=learning_rate,
    alpha=0.95,     # Decay rate for squared gradients
    momentum=0.95,   # Momentum for gradient updates
    eps=0.01         # Minimum squared gradient threshold
)
```

```python
# Modified Agent class
class Agent:
    def __init__(
        self,
        env: gym.Env,
        initial_epsilon: float,
        epsilon_decay: float,
        final_epsilon: float,
        discount_factor: float,
    ):
        self.env = env
        self.epsilon = initial_epsilon
        self.epsilon_decay = epsilon_decay
        self.final_epsilon = final_epsilon
        self.discount_factor = discount_factor

    def get_action(self, state: np.ndarray, model: nn.Module) -> int:
        """
        Returns the best action with probability (1 - epsilon)
        otherwise a random action with probability epsilon to ensure exploration.
        """
        if np.random.random() < self.epsilon:
            return self.env.action_space.sample()
        else:
            # Convert state to tensor and normalize
            state_tensor = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0) / 255.0
            with torch.no_grad():
                q_values = model(state_tensor)
            return q_values.argmax().item()

    def decay_epsilon(self):
        self.epsilon = max(self.final_epsilon, self.epsilon - self.epsilon_decay)
```

```python
for episode in tqdm(range(n_episodes), desc="Training episodes"):
    # Reset environment
    obs, info = env.reset()
    frame = preprocess(obs)
    
    # Initialize frame stack
    frame_stack = deque(maxlen=4)
    for _ in range(4):
        frame_stack.append(frame)
    state = np.stack(frame_stack, axis=0)
    
    done = False
    truncated = False
    episode_reward = 0
    avg_loss = 0
    step_count = 0
    
    while not (done or truncated):
        # Get action
        action = agent.get_action(state, online_net)
        
        # Execute action
        next_obs, reward, done, truncated, info = env.step(action)
        episode_reward += reward
        
        # Preprocess and update frame stack
        next_frame = preprocess(next_obs)
        frame_stack.append(next_frame)
        next_state = np.stack(frame_stack, axis=0)
        
        # Store transition
        replay_memory.push((state, action, reward, next_state, done))
        state = next_state
        
        # Only update if we have enough samples
        if len(replay_memory) > minibatch_count:
            # Sample minibatch
            minibatch = replay_memory.sample(minibatch_count)
            states, actions, rewards, next_states, dones = zip(*minibatch)
            
            # Convert to tensors and normalize
            states = torch.tensor(np.array(states), dtype=torch.float32, device=device) / 255.0
            next_states = torch.tensor(np.array(next_states), dtype=torch.float32, device=device) / 255.0
            actions = torch.tensor(actions, dtype=torch.long, device=device)
            rewards = torch.tensor(rewards, dtype=torch.float32, device=device)
            dones = torch.tensor(dones, dtype=torch.float32, device=device)
            
            # Compute current Q values
            current_q = online_net(states).gather(1, actions.unsqueeze(1)).squeeze(1)
            
            # Compute next Q values with target network
            with torch.no_grad():
                next_q = target_net(next_states).max(1)[0]
                target_q = rewards + discount_factor * next_q * (1 - dones)
            
            # Compute loss
            loss = F.mse_loss(current_q, target_q)
            avg_loss += loss.item()
            step_count += 1
            
            # Optimize the model
            optimizer.zero_grad()
            loss.backward()
            
            # Gradient clipping (optional but recommended)
            for param in online_net.parameters():
                param.grad.data.clamp_(-1, 1)
                
            optimizer.step()
        
        # Update target network periodically
        total_steps += 1
        if total_steps % target_update == 0:
            target_net.load_state_dict(online_net.state_dict())
    
    # Decay epsilon
    agent.decay_epsilon()
    
    # Record metrics
    episode_rewards.append(episode_reward)
    if step_count > 0:
        episode_losses.append(avg_loss / step_count)
    else:
        episode_losses.append(0)
    
    # Print progress
    if episode % 10 == 0:
        avg_reward = np.mean(episode_rewards[-10:])
        print(f"Episode {episode}: Reward={episode_reward:.2f}, Avg Reward (10)={avg_reward:.2f}, Epsilon={agent.epsilon:.4f}")
```
